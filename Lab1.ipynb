```python
def gradient_descent(x, lr):
    threshold = 1
    loss = []
    xs = []
    count = 0  
    print("Iter\t\tx\t\tf(x)")
    print("-" * 50)

    while x > threshold:
        f = x**2
        print(f"{count}\t\t{x:.6f}\t\t{f:.6f}")
        loss.append(f)
        xs.append(x)

        f_prime = 2 * x
        x = x - lr * f_prime  
        count += 1

    return loss

loss = gradient_descent(10, 0.1)
```

```python
import matplotlib.pyplot as plt

plt.plot(loss, marker='o')
plt.title("Loss Curve of Gradient Descent")
plt.xlabel("Iteration")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
